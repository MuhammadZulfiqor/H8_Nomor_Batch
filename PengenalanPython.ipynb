## Python Dictionary

Dictionary Python berbeda dengan List ataupun Tuple. Karena setiap urutanya berisi key dan value. Setiap key dipisahkan dari value-nya oleh titik dua (:), item dipisahkan oleh koma, dan semuanya tertutup dalam kurung kurawal. Dictionary kosong tanpa barang ditulis hanya dengan dua kurung kurawal, seperti ini: {}.

```python
MLB_team ={
    'Colorado': 'Rockies',
    'Boston' : 'Red Sox',
    'Minnesota': 'Twins',
    'Milwaukee': 'Brewers',
    'Seattle' : 'Mariners'
}
```

## Accesing Dictionary Values

sebuah nilai yang diambil dari dictionary dengan menentukan kunci yang sesuai dalam tanda kurung siku ([])

```python
print(MLB_team['Minnesota'])
print(MLB_team['Colorado'])
```

```python
#adding atau menambahkan new key dan value
MLB_team['Kabsas City'] = 'Royals'
MLB_team
```

```python
#mengupdate nilai value pada directory
MLB_team['Seattle'] = 'Seahawks'
MLB_team
```

```python
#mendelete nilai key dan value pada directory

del MLB_team['Seattle']
MLB_team
```

## Building a dictionary incrementally

mendefinisikan dictionary menggunakan tanda kurung kurawal dan key-value pairs, seperti yang di tunjukkan di atas

```python
person = {}
type(person)

person['fname']= 'Hack'
person['lname']= 'PTP'
person['age'] = 51
person['spouse'] = 'Edna'
person['children'] = ['Ralph', 'Betty', 'Joey']
person['pets'] = {'dog': 'fido','cat': 'Sox'}
```

```python
person
```

```python
print(person['fname'])
print(person['lname'])
```

```python
print(person['children'])
print(person['children'][1])
```

```python
print(person['pets'])
print(person['pets']['cat'])
```

```python
#built-in methods
d = {'a': 10, 'b': 20, 'c':30}

#items
print(d.items())

#keys
print(d.keys())

#values
print(d.values())
```

## Line Continuation

```python
person1_age = 42
person2_age = 16
person3_age = 71

someone_is_of_working_age = (person1_age >= 18 and person1_age <= 65) or (person2_age >= 18 and person2_age <= 65) or (person3_age >= 18 and person3_age <= 65)
someone_is_of_working_age
```

```python
someone_is_of_working_age = (
    (person1_age >= 18 and person1_age <= 65)
    or (person2_age >= 18 and person2_age <= 65)
    or (person3_age >= 18 and person3_age <= 65)
)
someone_is_of_working_age
```
